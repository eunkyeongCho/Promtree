```python
import os
import pymupdf

os.makedirs("pdf_images", exist_ok=True)

for msds in msds_list:
    path = msds.split('.')[0]
    os.makedirs(f"pdf_images/{path}", exist_ok=True)

    doc = pymupdf.open(f"../../data/msds/{msds}")

    for page_num in range(doc.page_count):
        page = doc[page_num]
        pix = page.get_pixmap()
        pix.save(f"pdf_images/{path}/page_{page_num + 1}.jpeg")
    
    doc.close()
```

```python
from paddleocr import LayoutDetection

model = LayoutDetection(model_name="PP-DocLayout_plus-L")
output = model.predict(r"pdf_images\5bc0c676-018f-46de-bb0d-0103ff9c388c\page_1.jpeg", batch_size=1, layout_nms=True)
for res in output:
    res.print()
    res.save_to_img(save_path="./output/")
    res.save_to_json(save_path="./output/res.json")
```

```python
import os
from paddleocr import LayoutDetection

model = LayoutDetection(model_name="PP-DocLayout_plus-L")

pdf_images = os.listdir("pdf_images")

for pdf_images_dir in pdf_images:
    pages_jpeg = os.listdir(f"pdf_images/{pdf_images_dir}")
    for image in pages_jpeg:
        output = model.predict(f"pdf_images/{pdf_images_dir}/{image}", batch_size=1, layout_nms=True)
        for res in output:
            res.save_to_img(save_path=f"./output/{pdf_images_dir}/{image}")
```
